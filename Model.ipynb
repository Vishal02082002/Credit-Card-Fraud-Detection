```python
import numpy as np
import pandas as pd
#import matplotlib.pyplot as plt
import operator
from sklearn.naive_bayes import MultinomialNB
# import matplotlib.pyplot as plt
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
# load the data set 
data = pd.read_csv('ccFraud.csv') 
ccdata=data.head(1000000) 
# print info about columns in the dataframe 
print(ccdata.info()) 
ccdata.head(10)
```

```python
# MODEL MAKING BEGINS - Seperating dependent and independent variables
# Dropping custId column as it autoincrements and no affect on results 
labels = ccdata.columns[1:8]
l1 = ccdata.columns[8]
X = ccdata[labels]
y = ccdata[l1]
print(X)
```

```python
# split into 70:30 ration 
X_trn, X_tst, y_trn, y_tst = train_test_split(X, y, test_size = 0.4, random_state = 5) 
mg = MultinomialNB()
mg.fit(X_trn, y_trn.ravel())
ypredmg = mg.predict(X_tst)
```

```python
print(ypredmg)
```

```python
print(confusion_matrix(y_tst,ypredmg))
print(classification_report(y_tst,ypredmg))
amg = accuracy_score(y_tst, ypredmg)
print("Accuracy: %.2f%%" % (amg* 100.0))
```

```python
print(confusion_matrix(y_tst,ypredmg))
print(classification_report(y_tst,ypredmg))
amg = accuracy_score(y_tst, ypredmg)
print("Accuracy: %.2f%%" % (amg* 100.0))
```
